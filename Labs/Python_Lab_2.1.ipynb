### Task 1
1. Write a function named is_prime that takes a number as a parameter and returns True if the number is prime and False otherwise. Print whether the number 29 is prime.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test the function with the number 29
number = 29
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

3. Use the zip() function to combine three lists (names, ages, and cities) into a list of tuples.
4. Use the range() function to create a list of even numbers from 2 to 20 and print the list.

### Task 3
Build a phone book program that receives the phone number, and returns the name of the owner.
You can follow the table below:

| Name     | Number     |
| -------- | ---------- |
| Amal     | 1111111111 |
| Mohammed | 2222222222 |
| Khadijah | 3333333333 |
| Abdullah | 4444444444 |
| Rawan    | 5555555555 |
| Faisal   | 6666666666 |
| Layla    | 7777777777 |

- If the number exists, print the owner. Otherwise, print "Sorry, the number is not found".
- If the number is less or more than 10 numbers, print "This is invalid number".
- If the number contains letters or symbols, print "This is invalid number".


```python
phone_book = {
    "Amal": "1111111111",
    "Mohammed": "2222222222",
    "Khadijah": "3333333333",
    "Abdullah": "4444444444",
    "Rawan": "5555555555",
    "Faisal": "6666666666",
    "Layla": "7777777777"
}

def find_owner(number):
    # Check if the number contains only digits and is 10 characters long
    if not number.isdigit():
        print("This is invalid number")
    elif len(number) != 10:
        print("This is invalid number")
    else:
        # Search for the owner by number
        for name, phone in phone_book.items():
            if phone == number:
                print(f"The owner is: {name}")
                return
        print("Sorry, the number is not found")

# Example usage
find_owner(input("Enter the phone number: "))
```

### Task 4
- Given the following list : [5, 4, 17, 19, 30, 2, 7, 10, 45]
    1. Use a lambda function to filter out only the odd numbers.
    2. Print the new list of odd numbers.

```python
numbers = [5, 4, 17, 19, 30, 2, 7, 10, 45]

# Use a lambda function to filter out only the odd numbers
odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))

# Print the new list of odd numbers
print("List of odd numbers:", odd_numbers)
```
